```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
thresh_lst = []
max_lst = []
for i in range(9, 17):
    thresh_file_path = "../data/statistics/20170417/{}/acc_thresh.csv".format(i)
    current_thresh_lst = np.loadtxt(thresh_file_path, delimiter=",")
    thresh_lst.extend(current_thresh_lst[:-120])
    
    max_file_path = "../data/statistics/20170417/{}/max.csv".format(i)
    current_max_lst = np.loadtxt(max_file_path, delimiter=",")
    max_lst.extend(current_max_lst[:-120])
```

```python
plt.figure(figsize=(16, 4))
plt.plot(thresh_lst, color="r")
plt.plot(max_lst)
plt.tick_params(labelsize=16)
plt.xticks([i*107900 for i in range(9)], ["9:00", "10:00", "11:00", "12:00", "13:00", "14:00", "15:00", "16:00", "17:00"])
plt.grid()
plt.ylim(0, 500)
```
