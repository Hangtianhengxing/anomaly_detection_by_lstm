```python
import gc
from tqdm import tqdm
```

```python
def window_datasets(dataset_df, col_lst, window_size=30, interval=5):
    for wnd_idx in range(1, window_size+1, interval):
        window_df = dataset_df.copy()
        window_df = window_df[col_lst]
        window_df = window_df.shift(wnd_idx)
        window_df = window_df.rename(columns=lambda s: "{0}_prev{1}".format(s, wnd_idx))
        dataset_df = dataset_df.join(window_df)
        del window_df
        gc.collect()
        
    dataset_df = dataset_df.drop([i for i in range(window_size)]).reset_index(drop=True)
    
    return dataset_df
```

```python
day_lst = ["20170416", "20170417", "20170420", "20170421", "20170422", "20181030", "20181031"]
window_size = 60*10
interval = 10
window_col_lst = ['degree_mean', 'degree_mean_shift1', 'degree_std', 'degree_std_shift1', 'diver_cur',
       'diver_day', 'down', 'grid_0_0', 'grid_0_1', 'grid_0_2', 'grid_0_3', 'grid_0_4', 'grid_0_5', 'grid_0_6', 'grid_0_7',
       'horizontal', 'human_area', 'left', 'left_down', 'left_up', 'max', 'max_index_grid_0_0',
       'max_index_grid_0_1', 'max_index_grid_0_2', 'max_index_grid_0_3',
       'max_index_grid_0_4', 'max_index_grid_0_5', 'max_index_grid_0_6',
       'max_index_grid_0_7', 'max_shift1', 'mean', 'mean_shift1', 'oblique',
       'overall_dir_down', 'overall_dir_left', 'overall_dir_left_down',
       'overall_dir_left_up', 'overall_dir_right', 'overall_dir_right_down',
       'overall_dir_right_up', 'overall_dir_up', 'right', 'right_down',
       'right_up', 'up', 'var', 'var_shift1', 'vertical']

dataset_df = pd.DataFrame()
col_lst = ['Fri', 'Mon', 'Sat', 'Sun', 'Thurs', 'Tue', 'Wed', 'degree_mean',
       'degree_mean_shift1', 'degree_std', 'degree_std_shift1', 'diver_cur',
       'diver_day', 'down', 'feed_cur', 'feed_day', 'grid_0_0', 'grid_0_1',
       'grid_0_2', 'grid_0_3', 'grid_0_4', 'grid_0_5', 'grid_0_6', 'grid_0_7',
       'horizontal', 'hour_10', 'hour_11', 'hour_12', 'hour_13', 'hour_14',
       'hour_15', 'hour_16', 'hour_9', 'human_area', 'label', 'left',
       'left_down', 'left_up', 'max', 'max_index_grid_0_0',
       'max_index_grid_0_1', 'max_index_grid_0_2', 'max_index_grid_0_3',
       'max_index_grid_0_4', 'max_index_grid_0_5', 'max_index_grid_0_6',
       'max_index_grid_0_7', 'max_shift1', 'mean', 'mean_shift1', 'oblique',
       'overall_dir_down', 'overall_dir_left', 'overall_dir_left_down',
       'overall_dir_left_up', 'overall_dir_right', 'overall_dir_right_down',
       'overall_dir_right_up', 'overall_dir_up', 'right', 'right_down',
       'right_up', 'up', 'var', 'var_shift1', 'vertical', 'prev_acc_cnt']

init_df = pd.DataFrame(columns=col_lst)

for day in tqdm(day_lst):
    tmp_dataset_df = pd.read_csv("../../data/datasets/default/{0}.csv".format(day))

    # check NaN count
    assert tmp_dataset_df.isnull().values.sum() == 0

    tmp_dataset_df = pd.concat([tmp_dataset_df, init_df])
    tmp_dataset_df = window_datasets(tmp_dataset_df, window_col_lst, window_size, interval)

    dataset_df = pd.concat([dataset_df, tmp_dataset_df]).reset_index(drop=True)

    # shuffle dataset
    # dataset_df = dataset_df.sample(frac=1)
    dataset_df = dataset_df.fillna(0)
    
    # check NaN count
    assert dataset_df.isnull().values.sum() == 0
    
for i in range(5):
    dataset_df.loc[dataset_df["label"] == i+1, "label"] = i
    
print(len(dataset_df))
dataset_df.head()
```
