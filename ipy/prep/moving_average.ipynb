```python
data_df = pd.DataFrame()

for i in range(9, 17):
    tmp_data_df = pd.read_csv("../data/datasets/20181031/default/time_series_{}.csv".format(i))
    data_df = pd.concat([data_df, tmp_data_df])

data_df = data_df.fillna(0)    
data_df.head()
```

```python
mean_arr = np.array(data_df["mean"])
plt.figure(figsize=(12, 4))
plt.plot(mean_arr)
```

```python
moving_size = 60*20
b = np.ones(moving_size)/moving_size
mean_ave = np.convolve(mean_arr, b, "same")

plt.figure(figsize=(12, 4))
plt.plot(mean_arr)
plt.plot(mean_ave)
plt.ylim(0, 60)
```
