```python
thresh_lst = []
max_lst = []
for i in range(9, 17):
#     thresh_file_path = "../../data/statistics/20181030/{}/acc_thresh.csv".format(i)
#     current_thresh_lst = list(pd.read_csv(thresh_file_path)["acc_thresh"])
#     thresh_lst.extend(current_thresh_lst[:-120])
    
    if i != 10:
        max_file_path = "../../data/statistics/20181031/{}/mean.csv".format(i)
        current_max_lst = list(pd.read_csv(max_file_path)["mean"])
        max_lst.extend(current_max_lst[:-120])
```
