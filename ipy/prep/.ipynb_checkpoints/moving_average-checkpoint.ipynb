```python
day = "20170422"
data_df = pd.DataFrame()

for i in range(9, 17):
    tmp_data_df = pd.read_csv("../../data/datasets/{}/default/time_series_{}.csv".format(day, i))
    data_df = pd.concat([data_df, tmp_data_df])

data_df = data_df.fillna(0)    
data_df.head()
```

```python
mean_arr = np.array(data_df["mean"])
plt.figure(figsize=(12, 4))
plt.plot(mean_arr)
```

```python
moving_size = 60*10
b = np.ones(moving_size)/moving_size
mean_ave = np.convolve(mean_arr, b, "same")

plt.figure(figsize=(12, 4))
plt.title("Mean ({})".format(day))
plt.plot(mean_arr)
plt.plot(mean_ave, label="moving_mean")
plt.plot(np.array(data_df["feed"])*120, label="feed")
plt.legend(loc="upper right")
plt.xticks([i*3594 for i in range(9)], ["9:00", "10:00", "11:00", "12:00", "13:00", "14:00", "15:00", "16:00", "17:00"])
plt.ylim(0, 120)
plt.ylabel("mean [px/sec]")
plt.xlabel("time")

plt.savefig("../../image/out_{}.png".format(day))
```
