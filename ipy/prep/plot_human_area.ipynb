```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
human_lst = []
for i in range(9, 17):
    file_path = "../data/output/human_area/20170416/human_{}.csv".format(i)
    human_arr = np.loadtxt(file_path)
    human_arr[human_arr >= 0.9] = 0
    human_lst.extend(list(human_arr[:-120]*100))
    
print(len(human_lst))
print(human_lst[:10])
```

```python
plt.figure(figsize=(16,4))
plt.plot(human_lst)
plt.tick_params(labelsize=16)
plt.title("水槽前方の人の密度")
plt.xlabel("時間")
plt.ylabel("密度 [%]")
plt.ylim(0, 100)
plt.xticks([i*107900 for i in range(9)], ["9:00", "10:00", "11:00", "12:00", "13:00", "14:00", "15:00", "16:00", "17:00"])
plt.grid()
```
