```python
def normalize(df, col_lst):
    result_df = df.copy()
    for col in col_lst:
        max_value = df[col].max()
        min_value = df[col].min()
        result_df[col] = (df[col] - min_value)/(max_value - min_value)
    
    return result_df
```

```python
col_lst = ["mean", "var", "max", "degree_mean", "degree_std", "max_count"]
normalized_df = normalize(dataset_df, col_lst)
normalized_df.head()
```
